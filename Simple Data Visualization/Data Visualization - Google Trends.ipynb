```python
df1 = df[df['query'] == 'Facebook']
df1.head()
```

```python
df1.year.value_counts()
```

Looks like the searches for Facebook vary over time. 

```python
f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)

sns.histplot(
    df1,
    x="year", hue="rank",
     multiple="stack",
    palette="rocket",
    edgecolor=".3",
    linewidth=.5
)
```

It sort of makes sense that Facebook was searched for a lot in the year 2009 - it appears that "Facebook announced that it had 150 million users in 2009, and it became mainstream."

But another question rise - what happen to year 2010?

```python
df2 = df[df['year']==2010]
df2['query'].value_counts()
```

```python
df2[df2['location']=='Global']
```

It looks like Netflix and YouTube were searched more that year (definitely surpassed FaceBook searches), which is not surprising. 

Now I want to take a look at the people searches. 

```python
df4 = df[df['category']=='People']
df4['query'].value_counts()
```

```python
array = ['Whitney Houston','Amy Winehouse','Adele']
df5 = df4.loc[df4['query'].isin(array)]
```

```python
df5['location'].value_counts()
```

```python
array2 = ['Czechia','United Kingdom', 'Poland']
df6 = df5.loc[df5['location'].isin(array2)]
```

```python
sns.displot(
    df6,x='query', col="location", row="year",
    facet_kws=dict(margin_titles=True)
)
```

In those three countries, Adele and Amy Winehouse were very popular in 2011, but in 2012, due to the tragedy of Whiteney Houston, she tops the google search for across all three selected countries.
